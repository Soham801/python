```python
# Data Serialization

import json
data = {
    'name':'Soham',
    'age':'21'
}

json_str = json.dumps(data) # Command for converting dictionary data to json data in string
print(json_str)
print(type(json_str)) # Understanding the type of data


parsed_data = json.loads(json_str) # This command converts json string data back to dictionary data 
print(parsed_data)
print(type(parsed_data)) # Undertanding the type of data
```

```python
## CSV

import csv

with open('example.csv',mode='w',newline='') as file: # Creating new CSV file in the folder in write mode ('w')
    writer=csv.writer(file) # Used to initialize editing in the particular file
    writer.writerow(['name','age']) # Creating a row in the CSV file
    writer.writerow(['Soham','21']) # Creating a row in the CSV file

with open('example.csv',mode='r') as file: # Allowing the particular CSV file to read
    reader=csv.reader(file) # Initializing the read in CSV file
    for row in reader: # Reading the rows in the CSV file 
        print(row)
```

```python
## Date Time
from datetime import datetime,timedelta

now = datetime.now() # Gives the current Date and time
print(now)

yesterday = now - timedelta(days=1) # Give the same exact date and time of yesterday , can increase the days
print(yesterday)
```

```python
# Time

import time
print(time.time())
time.sleep(2)
print(time.time())  # Makes the program sleep and re run in mentioned time duration
```

```python
# Regular expression

import re

pattern = r'\d+' # Pattern to find digits

text = "There are 123 apples in the garden ,456 mangoes and 12 bananas"
match = re.search(pattern,text) # Searches for the pattern in the above mentioned text

# When the pattern is first found the function wont print other patterns 

print(match.group()) # prints where the pattern was found in the text
```
