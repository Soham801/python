```python
# Data Serialization

import json
data = {
    'name':'Soham',
    'age':'21'
}

json_str = json.dumps(data) # Command for converting dictionary data to json data in string
print(json_str)
print(type(json_str)) # Understanding the type of data


parsed_data = json.loads(json_str) # This command converts json string data back to dictionary data 
print(parsed_data)
print(type(parsed_data)) # Undertanding the type of data
```

```python
## CSV

import csv

with open('example.csv',mode='w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['Soham','21'])

with open('example.csv',mode='r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```
